### This notebook allows the user to look at all the content that is not defined by the Evaluator service. The notebook uses the data prepared by the Evaluator service. It can be used to identify the improvements in the concept crosswalks the Evaluator is created from. 

Import the python packages needed

```python
import pandas as pd
import os
from os import walk
import fnmatch
from ipywidgets import *
import ipywidgets as widgets
pd.options.display.max_colwidth=200
```

Query the data directory for data files and use the results to populate a list

```python
DataFiles=[]
for path, subdirectories, filenames in os.walk('../data/'):
    for filename in filenames:
        if fnmatch.fnmatch(filename, '*.csv'):
            DataFiles.append(os.path.join(path,filename).split("../data/", 1)[-1])
DataFiles        
```

### Choose the data you want to examine. data.csv is the file created by the Evaluator in the Metadata to Data notebook

Function that reads the selected csv into a dataframe

```python
def DataChoices(DataFile):
    global CollectionConceptsDF
    CollectionConceptsDF= pd.read_csv(os.path.join('../data', DataFile))
    return CollectionConceptsDF
```

Choose the CSV you want to examine by using the list of csv in a dropdown menu and using the result in the DataChoices function.

```python
interactive(DataChoices, DataFile=DataFiles)
```

###  Let's get rid of the columns we won't use for this analysis. We already know the organization and dialect after we have selected the organization and collection. The xpaths aren't needed either.

Drop the columns listed from the dataframe

```python
CollectionConceptsDF.drop(['Record','Collection','Dialect'], axis=1, inplace=True)
CollectionConceptsDF
```

Create a slice of the dataframe that contains just rows where Unknown is in the Concept Column.

```python
UnknownTableDF = CollectionConceptsDF[CollectionConceptsDF.Concept == "Unknown"]
UnknownTableDF
```

## Select a specific XPath and look at all of the examples in the collection

Create a unique list of XPaths 

```python
XPathVerticals=UnknownTableDF.XPath.unique()
Verticals=XPathVerticals.tolist()
Verticals
```

Create a function that uses a specific XPath to create a new dataframe of just rows with the selected XPath.

```python
def ConceptVerticalTable(XPath):
    global VerticalTable
    VerticalTable = UnknownTableDF[UnknownTableDF.XPath == XPath]
    return VerticalTable
VerticalTable
```

Create a dropdown that selects from the list of unique XPaths and uses the result in the ConceptVerticalTable function

```python
interactive(ConceptVerticalTable,XPath=Verticals)
```
